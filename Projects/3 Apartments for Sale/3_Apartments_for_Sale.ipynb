# Исследование объявлений о продаже квартир

В вашем распоряжении данные сервиса Яндекc Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Ваша задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

<b>Задача</b> - определить влияние параметров на стоимость квартиры и определить те, на которые стоит ориентироваться при определении стоимости квартиры.
- В первой части изучим датасет и всё его содержимое, чтоб определить тот объём, на основе которого будем проводить поиск зависимостей. Пройдёмся по каждому столбцу и найдём аномалии значений, очевидные ошибки записи, посмотрим на типы данных, на пустоты, попробуем определить решение для каждой выявленной ситуации.
- Во второй части реализуем поставленные в первой задачи по приведению датасета в рабочее состояние.
- В третьей части произведём вычисления новых параметров и добавим их к таблице.
- В четвёртой части проведём анализ данных, очистим от выбросов, установим зависимости и подберём параметры, на которые можно будет опираться при определении стоимости квартиры.

### Описание данных

    airports_nearest — расстояние до ближайшего аэропорта в метрах (м)
    balcony — число балконов
    ceiling_height — высота потолков (м)
    cityCenters_nearest — расстояние до центра города (м)
    days_exposition — сколько дней было размещено объявление (от публикации до снятия)
    first_day_exposition — дата публикации
    floor — этаж
    floors_total — всего этажей в доме
    is_apartment — апартаменты (булев тип)
    kitchen_area — площадь кухни в квадратных метрах (м²)
    last_price — цена на момент снятия с публикации
    living_area — жилая площадь в квадратных метрах (м²)
    locality_name — название населённого пункта
    open_plan — свободная планировка (булев тип)
    parks_around3000 — число парков в радиусе 3 км
    parks_nearest — расстояние до ближайшего парка (м)
    ponds_around3000 — число водоёмов в радиусе 3 км
    ponds_nearest — расстояние до ближайшего водоёма (м)
    rooms — число комнат
    studio — квартира-студия (булев тип)
    total_area — площадь квартиры в квадратных метрах (м²)
    total_images — число фотографий квартиры в объявлении

<a id='название_осей'>Функция для подписи осей на гистограмме</a>

```python
ax = df.hist(column='total_area', bins=80, range=(0, 400),figsize=(15,5), grid=True)

#---------------название осей------------------------------------------------
def histogram(ax, x_label, y_label):

    ax = ax[0]
    for x in ax:
    
        # Remove title
        x.set_title("")
    
        # Set x-axis label
        x.set_xlabel(x_label, labelpad=20, weight='bold', size=12)

        # Set y-axis label
        x.set_ylabel(y_label, labelpad=20, weight='bold', size=12)
#---------------------------------------------------------------------------

histogram(ax, "Total Area (square meter)", "Quantity")
```

Функция для обработки выбросов<a id='обработать_выбросы'></a>

```python
# остаётся диапазон средних значений для исключения выбросов
def middle(df, col_name):
    q3 = df[col_name].quantile(0.75)
    q1 = df[col_name].quantile(0.25)
    min_itm = q1-(q3 - q1)*1.5
    max_itm = q3 + (q3 - q1)*1.5
    print('до', len(df))
    df = df[(df[col_name] > min_itm) & (df[col_name] < max_itm)]
    print('после', len(df))
    return(df)
    
df = middle(df, 'total_area')
```
